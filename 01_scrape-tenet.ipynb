```python
from youtube_transcript_api import YouTubeTranscriptApi
import requests
from bs4 import BeautifulSoup
import pandas as pd

def get_video_id_from_url(url):
    """Extract the video ID from a YouTube URL."""
    if "watch?v=" in url:
        return url.split("watch?v=")[-1]
    elif "youtu.be/" in url:
        return url.split("youtu.be/")[-1]
    return None

def get_video_title(video_id):
    """Fetch the video title by scraping the YouTube video page."""
    url = f"https://www.youtube.com/watch?v={video_id}"
    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'html.parser')
    
    # Extract the title tag content
    title = soup.find("meta", property="og:title")['content']
    return title

def get_captions_and_title(video_id):
    """Fetch the captions and title for a YouTube video."""
    try:
        # Fetch captions
        transcript = YouTubeTranscriptApi.get_transcript(video_id)
        captions = " ".join([entry['text'] for entry in transcript])  # Join all captions into a single string

        # Fetch video title
        title = get_video_title(video_id)
        
        return {
            'title': title,
            'captions': captions
        }
    except Exception as e:
        print(f"Error fetching data for video {video_id}: {e}")
        return None

def process_videos(video_urls):
    """Iterate through a list of YouTube video URLs and extract the transcript and title."""
    data = []
    
    for url in video_urls:
        video_id = get_video_id_from_url(url)
        if video_id:
            print(f"Processing video ID: {video_id}")
            video_data = get_captions_and_title(video_id)
            if video_data:
                data.append({
                    'Video URL': url,
                    'Video Title': video_data['title'],
                    'Transcript': video_data['captions']
                })
    
    # Convert data into a Pandas DataFrame
    df = pd.DataFrame(data)
    return df


# Process videos and extract data
df = process_videos(video_urls)

# Display the DataFrame
print(df)

# Save the DataFrame to a CSV file (optional)
df.to_csv('youtube_video_data.csv', index=False)
```
